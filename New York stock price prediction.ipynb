```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['symbol'] = le.fit_transform(df['symbol'])
```

```python
num_features = []
for dtype,feature in zip(df.dtypes,df.columns):
    if dtype == 'float64' or dtype == 'int64':
        num_features.append(feature)
```

```python
from sklearn.model_selection import train_test_split
```

```python
df = df[num_features]
```

```python
y_train = df['close']
```

```python
df = df.drop(['close'],axis=1)
```

```python
x_train,x_test,train_y,test_y = train_test_split(df,y_train,test_size = 0.25,random_state = 42)
```

```python
x_test
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,train_y)
```

```python
from sklearn.metrics import accuracy_score 
y_prediction = lr.predict(x_test)
```

```python
y_prediction
```

```python
test_y
```

```python
acc = accuracy_score
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
mse_multiple = mean_squared_error( y_prediction,test_y)
print("MSE of multiple linear regression: {:.0f}".format(mse_multiple))
```

```python
r2 = metrics.r2_score(y_prediction, test_y) # 使用 r-square 評估
print(r2)
```
